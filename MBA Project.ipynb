```python
#import necessary libraries
import pandas as pd
import json
```

# Processing Match Results

```python
with open("C:/Users/Prasana/OneDrive/Desktop/2023 worldcup Match Summary.json") as f:
    data = json.load(f)
```

```python
df_match = pd.DataFrame(data[0]['matchSummary'])
df_match.head()
```

```python
df_match.shape
```

```python
## Converting the column name[Scorecard] to match_id to connect with other tables
```

```python
df_match.rename({'scorecard': 'match_id'}, axis = 1, inplace = True)
df_match.head()
```

```python
## Creating a Dictionary to link with other tables
```

```python
match_ids_dict = {}

for index, row in df_match.iterrows():
    key1 = row['team1'] + ' Vs ' + row['team2']
    key2 = row['team2'] + ' Vs ' + row['team1']
    match_ids_dict[key1] = row['match_id']
    match_ids_dict[key2] = row['match_id']
```

```python
df_match.to_csv("D:/MBA/Files/match_summary.csv",index= False)
```

# Processing Batting Summary

```python
df_batting=pd.read_csv("D:/data/batting_summary.csv")
```

```python
df_batting.head(12)
```

```python
df_batting.isnull().sum()
```

```python
df_batting.fillna('no', inplace=True)
```

```python
df_batting['Dimissal_status'] = df_batting.dismissal.apply(lambda x: "out" if len(x)>3 else "notout")
```

```python
df_batting.head(12)
```

```python
df_batting.drop(columns=["dismissal"], inplace=True)
df_batting.head(10)
```

```python
df_batting['match_id'] = df_batting['match'].map(match_ids_dict)
```

```python
df_batting.tail()
```

```python
df_batting['batsmanName'] = df_batting['batsmanName'].apply(lambda x: x.replace('â€', ''))
df_batting['batsmanName'] = df_batting['batsmanName'].apply(lambda x: x.replace('\xa0', ''))
df_batting.head()
```

```python
df_batting.to_csv("D:/MBA/Files/batting_summary.csv", index = False)
```
